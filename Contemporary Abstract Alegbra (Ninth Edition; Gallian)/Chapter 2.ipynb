> **c.** function composition of polynomials with real coefficients
>> Associative. Let $f, g, h: \mathbb{R} \rightarrow \mathbb{R}$ be polynomial functions with real coefficients. For any $x \in \mathbb{R}$, we have
>> $$(f \circ (g \circ h))(x) = f((g \circ h)(x)) = f(g(h(x))) = (f \circ g)(h(x)) = ((f \circ g) \circ h)(x).$$

## Exercise 14

For group elements $a$, $b$, and $c$, express $(ab)^3$ and $(ab^{-2}c)^{-2}$ without parentheses.

For $(ab)^3$, we have

$$(ab)^3 = (ab)(ab)(ab) = ababab$$

(Note that associativity allows us to omit parantheses without risking ambiguity.)

For $(ab^{-2}c)^{-2}$, note that

$$(ab^{-2}c)^{-2} = \left((ab^{-2}c)^{-1}\right)^2.$$

Let $d = (ab^{-2}c)$. To find $d^{-1}$, recall that $dd^{-1} = d^{-1}d = e$, where $e$ is the identity. Choosing $d^{-1} = (c^{-1}b^2a^{-1})$, we have

$$dd^{-1} = (ab^{-2}c)(c^{-1}b^2a^{-1}) = ab^{-2}cc^{-1}b^2a^{-1} = e$$
$$d^{-1}d = (c^{-1}b^2a^{-1})(ab^{-2}c) = c^{-1}b^2a^{-1}ab^{-2}c = e$$

We also know $d^{-1}$ is unique. We then have

$$(ab^{-2}c)^{-2} = \left((ab^{-2}c)^{-1}\right)^2 = (c^{-1}b^2a^{-1})^2 = c^{-1}b^2a^{-1}c^{-1}b^2a^{-1}$$
